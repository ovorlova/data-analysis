```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
data = pd.read_csv("AirQualityUCI.csv", sep = ";", na_values = -200)
```

```python
data.drop('NMHC(GT)', axis = 1, inplace=True)
```

```python
data.info()
```

```python
data.plot() # data['T'] to choose collumn
data.plot(y = ['T', 'RH', 'AH'])
data.plot(x = 'T', y = ['RH', 'AH']) # only single collumn in x
```

```python
data.plot(x = 'Date', kind = 'line', figsize=(15,10), subplots=True, layout=(5,3))
#layout - format of output
```

```python
data.plot(x = 'T', y = 'AH', kind = 'scatter', figsize=(15,10)) 
# 'scatter' - without lines, only dots
#figsize - ratio of witdt and height
```

```python
data.plot(x = 'Date',y='T', kind = 'line', figsize=(15,10), subplots=True, title='Name')
#trying to make it
```

```python
plt.figure(figsize=(15,10))
ax = plt.plot(data['Date'], data['T'], 'c--^')
plt.legend(ax, ['T'])
plt.title('Name')
plt.show()
```

```python
data.rolling(window=200, min_periods=2).mean().plot(y='T', kind = 'line', figsize=(15,10), subplots=True, title='Name', style='r--')
```

```python
data.rolling(window=200, min_periods=2).mean().plot(kind = 'line', figsize=(15,10), subplots=True, title='Name', style=['r--'])
```

```python
data.plot(y = 'T',kind = 'hist', figsize=(15,10), subplots=True, title='Name',bins =25)
#Histogram
```

```python
data.plot(y = 'T',kind = 'box', figsize=(15,10), subplots=True, title='Name')
#boxPlot
```

```python
data.plot(y = 'T',kind = 'density', figsize=(15,10), subplots=True, title='Name')
#like histogram
```
