```python
for i in range(1, 14): 
    print(i,':\t', len(dataset[dataset['class'] == i]), sep='')
```

Пусть это будет группа 5:

```python
data = dataset[dataset['class'] == 5]
data
```

Уберем ненужные параметры **instr** и **class** - они не меняются
